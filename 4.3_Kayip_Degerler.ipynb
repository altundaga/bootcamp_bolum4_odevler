```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import warnings

warnings.filterwarnings('ignore')
#pembe ekran uyarıları için
```

```python
df=pd.read_csv('states_all.csv')
```

```python
#soru1 
#degiskenlerin tiplerini asagida görebiliyoruz
df.info()
```

```python
#her sutundaki kayıp değerlerin oranı yüzde olarak
df.isnull().sum()*100/df.shape[0]
```

```python
columns=['ENROLL','TOTAL_REVENUE','FEDERAL_REVENUE','STATE_REVENUE','LOCAL_REVENUE','TOTAL_EXPENDITURE','INSTRUCTION_EXPENDITURE','SUPPORT_SERVICES_EXPENDITURE','OTHER_EXPENDITURE','CAPITAL_OUTLAY_EXPENDITURE','GRADES_PK_G','GRADES_KG_G','GRADES_4_G','GRADES_8_G','GRADES_12_G','GRADES_1_8_G','GRADES_9_12_G','GRADES_ALL_G','AVG_MATH_4_SCORE','AVG_MATH_8_SCORE','AVG_READING_4_SCORE','AVG_READING_8_SCORE'] 
```

```python
#soru2
#kayıp veri olan sütunları kendi mean ile doldurduk
for i in columns:
    df_dolu_ort=df.fillna(df[i].mean())
#yılları dikkate almaksızın direkt ortalama ile doldurdugumuz veri cercevesini ayrı tuttuk.Sonradan kıyas yapabilmek için.
```

```python
df_dolu_ort.isnull().sum()
```

```python
#soru4
#yıl bazında ortalamardan önce interpolate ile eksik verileri doldurup yeni bir veri cercevesine kaydediyoruz.
df_polate=df.interpolate()
```

```python
#soru3
#bu sefer yılları dikkate alarak o yılın verilerinin  ortalamalarıyla dolduruyoruz eksik verileri.
for i in columns:
     df[i] = df.groupby("YEAR").transform(lambda x: x.fillna(x.mean()))
```

```python
df_yıl_ort=df
```

```python
# soru5 
#birbirinden farklı eksik veriyi doldurma yöntemlerini kıyaslamalıyım 
#ancak burada henüz bilimsel bir yol olan t-testi ve benzeri yöntemler işlenmedi.Kıyaslamak için yapabileceğim nedir?
#temel istatistiklerine bakıyorum belki buradan birşey anlaşılabilir bariz bir fark var ise
print(df_dolu_ort.describe(),'\n',df_yıl_ort.describe(),'\n',df_polate.describe())
```
