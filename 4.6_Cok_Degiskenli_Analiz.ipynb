```python
#ogle yemeginin ortalamardaki etkisi görülüyor ama yine de resmi olarak bu farkın anlamlı olup olmadıgına t-test ile bakıyoruz.
#bu matematik dersi için 
standard= df[df['lunch']=='standard']['math score']
free = df[df['lunch']=='free/reduced']['math score']

stats.ttest_ind(standard,free)
```

Standart ogle yemegi yiyen öğrencilerin tüm derslerde daha başarılı olduğu görülüyor.

```python
#soru3
#hazırlık kurslarının sınav performansı üzerinde etkisi
df.groupby('test preparation course').describe()
```

```python
df.groupby('test preparation course').mean()
```

```python
#Hazırlık kursunu bitiren öğrencilerin ders bazında ortalamalarının daha yüksek olduğu görülüyor.
#Yine testimizle  matematik dersi için bunun doğruluğuna bakalım
completed= df[df['test preparation course']=='completed']['math score']
none = df[df['test preparation course']=='none']['math score']

stats.ttest_ind(completed,none)
```
